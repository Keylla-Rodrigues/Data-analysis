```python
#1. Importar a tabela .csv

import pandas as pd
import numpy as np

df = pd.read_csv("Business_Licenses.csv")
display(df)
```

```python
df.columns
list(df.columns)
```

```python
# A tabela tem 7019 linhas e 27 colunas. Analisando a lista de colunas nesta tabela, varias colunas podem ser deletadas. 
# O proximo passo deleta-las para reduzir a tabela. 
```

```python
# Deletar as colunas q nao necessarias

df = df.drop(["X", "Y", "FID", "PropertyNumber","Business_Name","Civic_Unit", "Civic_House", "Civic_Street", "Mailing_Address","Detail", "GISLink", "Phone", "NAICS_Primary","Category","Permitted_Use", "Permitted_Use_Desc", "Latitude", "Longitude"], axis = 1) 
display(df)
```

```python
#Apos a deletion, restaram 9 colunas na tabela.  
```

```python
# info          # forma de verificar celulas vazias - imprimira um resumo de qtos valores nao nulos estao na tabela
                # essa forma tem que contar o complementar para saber qtas estao vazias. Ai entra a outra forma df.isna

print(df.info())     
```

```python
# A tabela acima (info) mostra a qty de celulas vazias para cada coluna. 
#Ex.: P/coluna "CommunityName" -> total entries(= 7019) menos 4209 non-null = 2810 nao nulos
#Ex.: P/coluna "Licence_Desc" -> total entries(= 7019) menos 7019 non-null = zero celulas vazias

# Essa "info" tem a vantagem de descrever o tipo de variavel: 
# Ex que chama a atencao "Approved_Date" esta sendo reconhecido como object, mas deveria ser data
```

```python
# df.isna().any()    or .sum()               # outra forma de verificar celulas vazias "isna -> eh vazio"

df.isna().any()
print(df.isna())
```
