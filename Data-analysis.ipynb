```python
#1. Importar a tabela .csv

import pandas as pd
import numpy as np

df = pd.read_csv("Business_Licenses.csv")
display(df)
```

```python
df.columns   # lista todas as colunas
list(df.columns)
```

##### A tabela tem 7019 linhas e 27 colunas. Analisando a lista de colunas nesta tabela, varias colunas podem ser deletadas
##### O proximo passo deleta-las para reduzir a tabela.

```python
df.dtypes  # aqui eh possivel ver como o Python categoriza o tipo de variavel. Approved_Date precisara ser trabalhado. 
```

```python
# info          # forma de verificar celulas vazias - imprimira um resumo de qtos valores nao nulos estao na tabela
                # essa forma tem que contar o complementar para saber qtas estao vazias. Ai entra a outra forma df.isna

print(df.info()) 
```

##### Alterando o tipo de variavel Object to data --> Approved_Date

## ANALISE DOS DADOS 

```python
# 1.  Quantas empresas foram abertas ao todo?

freq_bus_name = df["Business_Name"].count()
display(freq_bus_name)  
```

```python
# 2.  Quantas empresas foram abertas por area/local?

freq_bus_area = df["CommunityName"].value_counts()
display(freq_bus_area)
```

```python
# 3. Quais sao os top 3 por tipo de empresas que mais pediram abertura de empresa?
freq_type_bus = df["Licence_Desc"].value_counts()
display(freq_type_bus)
                                                            # Construcao (residential & general) seguida de filme
```

```python
# 4. Quantos funcionarios estao empregados (ao todo)?

total = round((df["Seasonal_Employees"].sum() + df["PartTime_Employees"].sum() + df["FullTime_Employees"].sum()))
print(total)
```

```python
# 5. Quantos funcionarios estao trabalhando por tipo de empregabilidade?
freq_seas_emp = round(df["Seasonal_Employees"].sum())
display(freq_seas_emp)

freq_part_emp = round(df["PartTime_Employees"].sum())
display(freq_part_emp)

freq_full_emp = round(df["FullTime_Employees"].sum())
display(freq_full_emp)

```
